### Нахождение минимумв функции метом золотого сечения

```python
import numpy as np

import math

import matplotlib
import matplotlib.pyplot as plt
```

```python
def fff(x):
    return math.sin(20 * x) * math.cos(4 * x)
```

```python
eps = 1e-6
prop = (1+math.sqrt(5.0))/2.0
```

```python
def zzz(a,b):
    
    h = math.fabs(fff(a) - fff(b)) 
    
    while h > eps:
        
        x2 = a + abs(b - a) / prop
        x1 = b - abs(b - a) / prop
        
        if (fff(x1)>=fff(x2)):
            a=x1
        else:
            b=x2
        h = abs(b - a)
        
    x = (a + b)/2.0
    return fff(x)
```

```python
a = 6.1
b = 6.4
zzz(a,b)
```

####  Тут я снова использую великолепную функцию Игоря из задания для порабол и все работает замечательно

```python
x = np.linspace(5.5, 6.5, 500)
Y = np.linspace(-1, 1, 100)
y = [fff(x) for x in x]
z = [zzz(a,b) for x in x]
A = [a for Y in Y]
B = [b for Y in Y]
#print(P)

fig = plt.figure()
fig.add_axes([0, 0, 1.5, 1.5])
plt.grid(True)

plt.plot(x, y)
plt.plot(x, z)
plt.plot(A, Y)
plt.plot(B, Y)
plt.show()
```
