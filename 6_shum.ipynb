```python
import numpy as np

import matplotlib
import matplotlib.pyplot as plt

from scipy.optimize import curve_fit
```

```python
def fff(x, a, b, c):
    return c / (1 + ((x - a)/b)**2 )
```

```python
a = 0.5
b = 1
c = 2
```

```python
x = np.arange(-2, 2, 0.1)

np.random.seed(123)
noise = 0.5 * np.random.normal(0, 0.2, x.size)
```

```python
n = fff(x, 0.5, 1, 2) + noise
y = fff(x, a, b, c)

fig = plt.figure()
fig.add_axes([0, 0, 1.5, 1.5])

plt.plot(x, y, label=r'эталонная функция')
plt.plot(x, n, 'o', label=r'данные с шумами')

plt.legend(loc='best')
plt.show()
```

```python
popt, ppp = curve_fit(fff, x, n, [1.0, 1.0, 1.0])
```

```python
fig = plt.figure()
fig.add_axes([0, 0, 1.5, 1.5])

plt.plot(x, fff(x,*popt), label=r'восстановленная функция')
plt.plot(x, n, 'o', label=r'данные с шумами')
plt.plot(x, y,'--', label=r'эталонная функция')

plt.legend(loc='best')
plt.show()
```
