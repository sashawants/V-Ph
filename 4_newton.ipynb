### –§—Ä–∞–∫—Ç–∞–ª –ù—é—Ç–æ–Ω–∞ –¥–ª—è —Ñ—É–Ω–∫—Ü–∏–∏ $ —Ö^3 = 1 $

```python
import matplotlib.pyplot as plt
%matplotlib inline

import math
import numpy as np
```

–ò—Å–ø–æ–ª—å–∑—É–µ–º—ã–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ:

```python
ep = 10**(-3)
#R = [1.0,1j*math.exp(math.pi/3.0),-math.pi/3.0]
R = [1.0,-(-1)**(1/3),(-1)**(2/3)]
x = np.arange(-1, 1, 0.001)
y = np.arange(-1, 1, 0.001)
xR = []
yR = []
xG = []
yG = []
xY = []
yY = []
```

–§—É–Ω—Ü–∏–∏:

```python
def fff(x):
    return x**3-1

def ddd(x):
    return 3*x**2

def iii(x):
    ki = -1
    ch = [0,0,0]
    while (sum(ch)==0):
        x = (( ddd(x) * x - fff(x)) / ddd(x))
        for i in range(3):
            if (abs(x-R[i])<ep):
                ch[i] = 1
        ki += 1
        nr = ch.index(max(ch))
        
    return (x,ki,nr)
    
```

–ü–æ—Å—Ç—Ä–æ–∏–º —Ñ—Ä–∞–∫—Ç–∞–ª—ã –ù—å—é—Ç–æ–Ω–∞:

üíî –¥–ª—è –∫–æ—Ä–Ω–µ–π -> $ 1 $
 
&#x1F49A;  –¥–ª—è –∫–æ—Ä–Ω–µ–π ->  $ e^{pi/2} $
 
&#x1F49B;  –¥–ª—è –∫–æ—Ä–Ω–µ–π -> $ -{pi/2} $

```python
for Re in x:
    for Im in y:
        
        z,ki,nr=iii(Re + 1j*Im)
        
        if (nr==0):
            xR.append(Re)
            yR.append(Im)
        elif (nr==1):
            xY.append(Re)
            yY.append(Im)
        else:
            xG.append(Re)
            yG.append(Im)
            
plt.figure(figsize=(15,10))
plt.plot(xG, yG, 'g,')
plt.plot(xY, yY, 'y,')
plt.plot(xR, yR, 'r,')
plt.show() 
```
